```python
import pandas as pd
import json
import os
```

```python
class Data_ingestion:
    def __init__(self, base_path="./"):
        self.data_url = os.path.join(base_path, '')
        self.data = None
        self.ingestion()
    
    def ingestion(self):
        csv_path = os.path.join(self.data_url, "PokemonGOData.csv")
        json_path = os.path.join(self.data_url, "pokemon_id_name.json")
        pokemon_go_data = pd.read_csv(csv_path)
        with open(json_path, 'r', encoding='utf-8') as f:
            pokemon_dict = json.load(f)
        self.data = [pokemon_go_data, pokemon_dict]

        
    def get_data(self):
        if not self.data:
            self.ingestion()
        return self.data
```
